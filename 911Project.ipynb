```python
def call_type_converter(x):
    x = x.split(':')
    return x[0]
```

```python
data['call_type'] = data['title'].apply(call_type_converter)
```

```python
data
```

```python
def emergency_seperator(x):
    x = x.split(':')
    return x[1]
```

```python
data['Emergency_Type'] = data['title'].apply(emergency_seperator)
```

```python
data.head()
```

```python
ems_count = data[data['call_type']=='EMS']['call_type'].value_counts()
Fire_count = data[data['call_type']=='Fire']['call_type'].value_counts()
Traffic_count = data[data['call_type']=='Traffic']['call_type'].value_counts()
```

```python
data_dict = {
    'EMS' : ems_count,
    'Fire' : Fire_count,
    'Traffic' : Traffic_count
}
```

```python
data_dict = pd.Series(data_dict)
```

### converting time stamp into pandas date and time

```python
data.head()
```

```python
data['timeStamp'] = pd.to_datetime(data['timeStamp'],infer_datetime_format=True)
```

```python
data.head()
```

```python
data.info
```

### Extracting year,month_name,day_name,hour of the day from time stamp 

```python
data['year'] = data['timeStamp'].dt.year
```

```python
data['month'] = data['timeStamp'].dt.month_name()
```

```python
data['day'] = data['timeStamp'].dt.day_name()
```

```python
data['hour'] = data['timeStamp'].dt.hour
```

```python
data.head()
```

```python
calls_data = data.groupby(['month'])['call_type'].value_counts()
```

```python
calls_data
```

```python
calls_data_percentage = calls_data.groupby(level=0).apply(lambda x : round((100*x) /float(x.sum())) )
```

```python
calls_data_percentage
```

```python
font = {
    'size': 'x-large',
    'weight': 'bold'
}

```

```python
sns.set(rc={'figure.figsize':(12, 8)})
calls_data_percentage.unstack().plot(kind='bar')
plt.xlabel('Name of the Month', fontdict=font)
plt.ylabel('Percentage of Calls', fontdict=font)
plt.xticks(rotation=0)
plt.title('Calls/Month', fontdict=font)
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
month_order
```

```python
calls_data_percentage = calls_data_percentage.reindex(month_order, level=0)
```

```python
calls_data_percentage = calls_data_percentage.reindex(['EMS','Traffic','Fire'], level=1)
```

```python
calls_data_percentage
```

```python
sns.set(rc={'figure.figsize':(12, 8)})
calls_data_percentage.unstack().plot(kind='bar')
plt.xlabel('Name of the Month', fontdict=font)
plt.ylabel('Percentage of Calls', fontdict=font)
plt.xticks(rotation=0)
plt.title('Calls/Month', fontdict=font)
```

### hourly data

```python
hourly_data = data.groupby(['hour'])['call_type'].value_counts()
```

```python
hourly_data
```

```python
hourly_data_percentage = hourly_data.groupby(level=0).apply(lambda x: round(100 * x/float(x.sum())))
```
